```python
import string
def decode_message(encoded_message, offset):
    decoded_list = []
    alphabelt = list(string.ascii_lowercase)
    for char in encoded_message:
        if char in alphabelt:
            temp_index = alphabelt.index(char) + offset
            if temp_index >= 26:
                decoded_index = temp_index % 26
            else:
                decoded_index = temp_index
            decoded_list.append(alphabelt[decoded_index])
        else:
            decoded_list.append(char)
    decoded_message = ''.join([str(n) for n in decoded_list])
    print(decoded_message)

decode_message("ebiil", 3)
```
